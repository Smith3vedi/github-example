```python
#import libraries
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.datasets.samples_generator import make_blobs

print("Libraries imported")
```

```python
#Read data into a dataframe
url = "https://cocl.us/customer_dataset"
customer_df = pd.read_csv(url)

print("Dataframe created")
```

```python
#to see the columns and get an idea
customer_df.head()
```

```python
customer_df.count()
```

```python
#Preprocessing/Cleaning data
#creating new dataframe by removing unnecessary information in the instance
#adress in categotical and not required
df = customer_df.drop('Address', axis =1)
df.head()
```

```python
#to normalize our numerical data
from sklearn.preprocessing import StandardScaler
x = df.values[:,1:]
x = np.nan_to_num(x) # assigning a numerical value to NAN
```

```python
cluster_dataset = StandardScaler().fit_transform(x)
cluster_dataset
```

```python
#modeling 
k = 3
k_means = KMeans(init = "k-means++", n_clusters = k , n_init =12)
k_means.fit(cluster_dataset)
labels = k_means.labels_

print(labels)
```

```python
#each row above is a customer

df["Labels"] = labels
df.head(10)
```

```python
df.groupby('Labels').mean()

##the labels are the centroid and the customer id gives an average approximiation on how many
    #customers belong to that particular centroid. 
    # The reamining columns are the average attributes speaking for the centroid
```
